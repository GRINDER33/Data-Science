```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5])
arr.dtype
```

```python
arr.astype('float64')
```

```python
arr2 = np.array([23, 4, 4, 60], dtype='float64')
arr2
```

```python
# DOWNCASTING TO SAVE SYSTEM MEMORY

arr_large = np.array([1000000, 2000000, 3000000], dtype='int64')
arr_small = arr_large.astype('int32')
print(arr_large)
print(arr_small.dtype)
```

```python
# MEMORY USAGE EX:

arr_int64 = np.array([1, 2, 3, 4], dtype='int64')
arr_int32 = np.array([1, 2, 3, 4], dtype='int32')

print(arr_int64.nbytes)  # 32 Bytes
print(arr_int32.nbytes)  # 16 Bytes
```

```python
# OBJECT DATA TYPE (NOT RECOMMENDED)

arr = np.array([{'a': 1}, [1, 2, 3], 'hello'], dtype='object')
print(arr)
```
