```python
import numpy as np
```

```python
arr = np.array([1, 2, 3, 4, 5, 25])
```

```python
# Compute the mean(Average) of an array
print(np.mean(arr))
```

```python
# Compute the minimum value of an array
print(np.min(arr))
```

```python
# Compute the maximum value of an array
print(np.max(arr))
```

```python
# Compute the Standard Deviation of an array 
print(np.std(arr))
```

```python
# Compute the variance of an array 
print(np.var(arr))
```

```python
# Compute the Sum of all elements in an array
print(np.sum(arr))
```

```python
# Compute the product of all elements in an array/
print(np.prod(arr))
```

```python
# Compute the median of an array
print(np.median(arr))
```

```python
# Compute the Percentile of an array
print(np.percentile(arr, 60))
```

```python
# Return the index of the minumum value in an array
print(np.argmin(arr))
```

```python
# Return the index of the maximum value in in array
print(np.argmax(arr))
```

```python
# Compute the correlation coefficiant matrix of two arrays
print(np.corrcoef(arr))
```

```python
# Find the unique elements of an array
print(np.unique(arr))
```

```python
# Compute the cumulative sum of an array
print(np.cumsum(arr))
```

```python
# Compute the n-th difference of an array
print(np.diff(arr))
```

```python
# Create an array with evenly spaced number over a specified interval
print(np.linspace(0, 10, 5))    # 5 numbers from 1 to 10 
```

```python
# Compute the natural logarithm of an array
print(np.log(arr))
```

```python
# Compute the exponential value of an array
print(np.exp(arr))
```
