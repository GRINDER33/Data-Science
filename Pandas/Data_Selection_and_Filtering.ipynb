```python
print(df[df['IMDb'] > 8.2])
```

```python
print(df[df['IMDb'] > 8.2]  [['Actor', 'Film']]  )  # Used to show  a specific column of those which qualify with the given parameter
```

```python
print(df[(df['IMDb'] > 8.2) & (df['Year'] > 2018)]  )  # | is 'or' and & is 'and'
```

```python
print(df.query("Year > 2019 and IMDb > 6"))   # .query() returns a copy not a view 
```

```python
df2 = df[['Actor', 'IMDb']].copy()   # It is not guaranteed for a copy use .copy() to confirm as a copy to make sure df is not modified
print(df2)
```
